A dataframes of the different LanguageWorkedWith status values in the dataset is store in language_used.This shows that most respondents have worked with more that one languages. This multiple answers by each respondent is not limited to only this column, we have the same occurence in the 3 other columns we will be using for our analysis, which mean they will all require cleaning.

```python
# This read the column for the language worked with.
language_used=data1['LanguageWorkedWith'].value_counts().reset_index()
language_used
```

Ploting subset of the language_used data, shows that we definitely need to clean this data. As shown in the plot below we need to seperate the different programing languages and count their respective numbers.

```python
# The distribution of the answer given by each of the respondent is shown as a bar chart
# This code takes time to run

plt.figure(figsize=(10,8))
plt.title('Programming Language Worked With ')
plt.xticks(rotation=90)
sns.barplot(x=language_used['index'][0:100],
            y=language_used['LanguageWorkedWith'][0:100]/sum(language_used['LanguageWorkedWith'])
            ,data=language_used,edgecolor='black')
```
